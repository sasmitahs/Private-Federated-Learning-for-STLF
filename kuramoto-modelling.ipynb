```python
import numpy as np
import pandas as pd
import scipy as sc
```

```python
def calc_phase(delta_t, delta_w_arr):

    theta_arr = []
    for i in range(len(delta_w_arr)):
        simmilarity = np.inner(delta_w_arr[i],delta_t)
        norm_t = np.linalg.norm(delta_t)
        norm_w = np.linalg.norm(delta_w_arr[i])

        denominator = norm_t*norm_w
        theta_i = simmilarity/denominator
        theta_arr.append(theta_i)
    
    return theta_arr
```

```python
def calc_sync_weights(phase_arr):

    theta_mean = np.sum(phase_arr)/len(phase_arr)

    base_sin = 0

    for i in range(len(phase_arr)):
        base_sin = base_sin + np.sin(theta_mean-phase_arr[i])

    sync_weights = []

    for i in range(len(phase_arr)):
        row_k = np.sin(theta_mean-phase_arr[i])
        client_sync_weight = row_k/base_sin
        sync_weights.append(client_sync_weight)
    
    return sync_weights
```
